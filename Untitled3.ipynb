```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Set up plot style
# plt.style.use("seaborn-darkgrid")

# Generate a synthetic dataset based on the variables provided
num_samples = 500
data = {
    "Trade Flows": np.random.choice(["Increasing", "Stable", "Decreasing"], num_samples),
    "Access to Finances": np.random.choice(["High", "Moderate", "Low"], num_samples),
    "Farming Type ": np.random.choice(["Communal", "Single-household", "Mixed"], num_samples),
    "Tariff Rates": np.random.randint(0, 20, num_samples),
    "Education and Skills": np.random.choice(["Advanced", "Basic", "Intermediate"], num_samples),
    "Employment Rates": np.random.randint(50, 90, num_samples),
    "Changes in Women's Income": np.random.choice(["Rising", "Falling", "Stable"], num_samples),
    "Business Ownership": np.random.randint(20, 150, num_samples),
    "Clear Decision Points": np.random.choice(["Yes", "No", "Partial"], num_samples),
    "Policy Changes": np.random.choice(["Significant", "Minor", "None"], num_samples),
    "Complex Interactions": np.random.choice(["High", "Medium", "Low"], num_samples),
    "Feedback Loops": np.random.choice(["Present", "Absent", "Weak"], num_samples),
    "Intra-African Mobility": np.random.choice(["Increasing", "Decreasing", "Stable"], num_samples),
    "Legal Frameworks": np.random.choice(["Supportive", "Neutral", "Restrictive"], num_samples),
    "Social Norms and Gender Roles": np.random.choice(["Progressive", "Traditional", "Mixed"], num_samples),
    "Women in Leadership": np.random.randint(0, 60, num_samples),
    "Access to Childcare": np.random.choice(["Good", "Limited", "None"], num_samples),
    "Impact on Women": np.random.choice(["Positive", "Negative", "Neutral"], num_samples),
    "Value Chain Participation": np.random.choice(["High", "Moderate", "Low"], num_samples),
    "Health and Well-being": np.random.choice(["Improved", "Declining", "Stable"], num_samples),
    "Women Empowerment Status": np.random.choice([1, 0], num_samples)
}

# Creating the DataFrame
df_large = pd.DataFrame(data)

df_large.head(15)

```

```python
# Step 1: Overview of the Data
print("Data Shape:", df_large.shape)
print("\nData Types:")
print(df_large.dtypes)
print("\nMissing Values Summary:")
print(df_large.isnull().sum())

print("\nBasic Statistics:")
print(df_large.describe(include="all"))
```

```python
# Step 2: Target Variable Analysis
plt.figure(figsize=(6, 4))
sns.countplot(data=df_large, x="Women Empowerment Status", palette="viridis")
plt.title("Distribution of Women Empowerment Status")
plt.xlabel("Empowerment Status")
plt.ylabel("Count")
plt.show()
```

```python
# Step 3: Univariate Analysis
categorical_features = df_large.select_dtypes(include="object").columns
plt.figure(figsize=(15, 25))
for i, col in enumerate(categorical_features, 1):
    plt.subplot(6, 3, i)
    sns.countplot(data=df_large, x=col, palette="viridis")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

numerical_features = df_large.select_dtypes(include=["int64", "float64"]).columns.drop("Women Empowerment Status")
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df_large[col], kde=True, palette="viridis")
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# Step 4: Bivariate Analysis
plt.figure(figsize=(15, 20))

# Calculate the number of rows needed for subplots based on the features
num_rows = int(np.ceil(len(categorical_features) / 3))

for i, col in enumerate(categorical_features, 1):
    plt.subplot(num_rows, 3, i)  # Use calculated num_rows
    sns.countplot(data=df_large, x=col, hue="Women Empowerment Status", palette="viridis")
    plt.title(f"{col} vs Women Empowerment Status")
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df_large, x="Women Empowerment Status", y=col, palette="viridis")
    plt.title(f"{col} vs Women Empowerment Status")
plt.tight_layout()
plt.show()
```

```python
# Step 5: Multivariate Analysis
plt.figure(figsize=(12, 8))
sns.heatmap(df_large[numerical_features].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

selected_features = ["Business Ownership", "Employment Rates", "Women in Leadership", "Tariff Rates"]
sns.pairplot(df_large, vars=selected_features, hue="Women Empowerment Status", palette="viridis")
plt.suptitle("Pairplot of Selected Numerical Features by Empowerment Status", y=1.02)
plt.show()
```

```python
# Step 6: Preprocessing for Modeling
df_encoded = df_large.copy()
label_encoders = {}
for column in df_encoded.select_dtypes(include="object").columns:
    label_encoders[column] = LabelEncoder()
    df_encoded[column] = label_encoders[column].fit_transform(df_encoded[column])

X = df_encoded.drop(columns="Women Empowerment Status")
y = df_encoded["Women Empowerment Status"]

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Step 7: Feature Engineering - Important Features Analysis
random_forest_feature_engineer = RandomForestClassifier(random_state=42)
random_forest_feature_engineer.fit(X_train, y_train)

feature_importances = random_forest_feature_engineer.feature_importances_
important_features = sorted(zip(feature_importances, df_encoded.columns[:-1]), reverse=True)
print("Top 10 Important Features:")
for importance, feature in important_features[:10]:
    print(f"{feature}: {importance:.4f}")
```

```python
# Import necessary metrics functions
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
# Step 8: Model Training and Evaluation
# Train Random Forest Model
random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train, y_train)
y_pred_rf = random_forest_model.predict(X_test)

# Train Logistic Regression Model
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, y_train)
y_pred_logistic = logistic_model.predict(X_test)

# Train K-Nearest Neighbors Model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

# Calculate Metrics for Each Model
models = {
    "Random Forest": y_pred_rf,
    "Logistic Regression": y_pred_logistic,
    "K-Nearest Neighbors": y_pred_knn
}

accuracy = {}
precision = {}
recall = {}
f1 = {}

for model_name, y_pred in models.items():
    accuracy[model_name] = accuracy_score(y_test, y_pred)
    precision[model_name] = precision_score(y_test, y_pred)
    recall[model_name] = recall_score(y_test, y_pred)
    f1[model_name] = f1_score(y_test, y_pred)

# Plot Confusion Matrices
# Plot Confusion Matrices with custom labels
plt.figure(figsize=(18, 5))
for i, (model_name, y_pred) in enumerate(models.items(), 1):
    cm = confusion_matrix(y_test, y_pred)
    plt.subplot(1, 3, i)
    sns.heatmap(cm, annot=True, fmt="d", cmap="viridis", cbar=False, annot_kws={"size": 14},
                xticklabels=["Women Not Empowered", "Women Empowered"],
                yticklabels=["Women Not Empowered", "Women Empowered"])
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# Plotting the Metrics for Each Model
metrics_df = pd.DataFrame({
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1
})

metrics_df.plot(kind="bar", figsize=(10, 6), colormap="viridis")
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.legend(loc="upper left")
plt.xticks(rotation=0)
plt.show()
```
